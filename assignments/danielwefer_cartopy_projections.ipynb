In this notebook I will be demonstrating various Cartopy projections by simply plotting coastlines

```python
import matplotlib.pyplot as plt       #basic plotting library

from cartopy import crs as ccrs       #Coordinate Reference System

from cartopy import feature as cfeature #cartopy's library of cartographic features

import warnings 

warnings.filterwarnings("ignore")       #filters warnings for visual purposes
```

## Plate Carree

While not a good projection for distortion reasons, Plate Carree is used as a default coordinate system for many datasets

```python
figure = plt.figure()
ax = plt.subplot(1, 1, 1, projection = ccrs.PlateCarree())
ax.coastlines()
plt.show()
```

### Albers Equal Area

##### While a little strange looking as a global view, Albers Equal Area is strong at preserving area in the midlatitudes

```python
figure = plt.figure()
ax = plt.subplot(1, 1, 1, projection = ccrs.AlbersEqualArea())
ax.coastlines()
plt.show()
```

### Azimuthal Equidistant

##### preserves both distance and direction from the central point

```python
figure = plt.figure()
ax = plt.subplot(1, 1, 1, projection = ccrs.AzimuthalEquidistant())
ax.coastlines()
plt.show()
```

### Eckert VI

##### An Equal Area projection, does not preserve shape, distance, direction, or angles

```python
figure = plt.figure()
ax = plt.subplot(1, 1, 1, projection = ccrs.EckertVI())
ax.coastlines()
plt.show()
```

### Equal Earth

##### An equal-area pseudocylindrical map projection, featuring curved edges to suggest Earth's curvature and straight parallels. 

```python
figure = plt.figure()
ax = plt.subplot(1, 1, 1, projection = ccrs.EqualEarth())
ax.coastlines()
plt.show()
```

## Equidistant Conic

#### A conic projection that preserves distance along all meridions and two standard parallels

```python
figure = plt.figure()
ax = plt.subplot(1, 1, 1, projection = ccrs.EquidistantConic())
ax.coastlines()
plt.show()
```

### Gnomonic

##### A Map projection that shows a sphere on a flat surface, often used in photography under the name rectilinear projection

```python
figure = plt.figure()
ax = plt.subplot(1, 1, 1, projection = ccrs.Gnomonic())
ax.coastlines()
plt.show()
```

## Lambert Conformal 

##### A map projection thta uses a Cone to represent the earth

```python
figure = plt.figure()
ax = plt.subplot(1, 1, 1, projection = ccrs.LambertConformal())
ax.coastlines()
plt.show()
```

## Lambert Cylindrical

##### An Equal-Area projection that is undistorted along the equator, but becomes very distorted poleward

```python
figure = plt.figure()
ax = plt.subplot(1, 1, 1, projection = ccrs.LambertCylindrical())
ax.coastlines()
plt.show()
```

### Mercator Projection

##### Used in classrooms around the world, this is a terrible projection for maintaining shape and scale, especially towards the poles. 

```python
figure = plt.figure()
ax = plt.subplot(1, 1, 1, projection = ccrs.Mercator())
ax.coastlines()
plt.show()
```

## Nearside Perspective

##### A projection that simulates the view of earth one would have from a satellite in orbit

```python
figure = plt.figure()
ax = plt.subplot(1, 1, 1, projection = ccrs.NearsidePerspective())
ax.coastlines()
plt.show()
```

## Interrupted Goode Homolosine

##### An equal-area projection intended for making world maps. The lower latitudes are comprised of six different Sinusoidal projections, and the higher latitudes are comprised of six separate Mollweide (homolographic) projections.

```python
figure = plt.figure()
ax = plt.subplot(1, 1, 1, projection = ccrs.InterruptedGoodeHomolosine())
ax.coastlines()
plt.show()
```

## Transverse Mercator

#### A projection similar to Mercator, however it projects along a meridian instead of the equator

```python
figure = plt.figure()
ax = plt.subplot(1, 1, 1, projection = ccrs.TransverseMercator())
ax.coastlines()
plt.show()
```
